### Identifica o tipo de transformação que permite obter o gráfico:

**1. De $f_2$ a partir do gráfico de $f_1$**

**2. De $f_3$ a partir do gráfico de $f_2$**

```python
from IPython.display import HTML
from base64 import b64encode

def play(GraficoAnimado):
    html = ''
    video = open(GraficoAnimado,'rb').read()
    src = 'data:video/mp4;base64,' + b64encode(video).decode()
    html += '<video width=1000 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src 
    return HTML(html)

play('GraficoAnimado.mp4')
```
