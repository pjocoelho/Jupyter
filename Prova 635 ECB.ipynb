```python
import pandas as pd
tabela = pd.read_excel("635_TURMAS.xlsx")
display(tabela)
```

```python
tabela[["Escrita"]].describe()
```

## Trinta melhores alunos

```python
tabela2 = tabela[["Nome", "Escrita", "Turma"]]
tabela2.sort_values(by=['Escrita'], ascending=False).head(30)
```

```python
turma_A = tabela.loc[tabela["Turma"] == "A", ["Nome", "Escrita"]]
display(turma_A)
```

```python
turma_A[["Escrita", "Classif"]].describe()
```

```python
turma_A = tabela.loc[tabela["Turma"] == "A", ["Nome", "Escrita"]]
turma_A.sort_values(by=['Escrita'], ascending=False).head(14)
```
